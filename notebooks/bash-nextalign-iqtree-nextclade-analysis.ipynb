Nextalign MSA of QC WTD SARS-CoV-2 sequences along with related sequences from GISAID and NCBI identified by UShER.

```bash
nextalign --version
```

```bash
nextalign run -r MN908947.3.fa \
  -o 2023-03-29-QC-WTD-and-related-GISAID-NCBI-seqs.nextalign.fasta \
  --include-reference \
  2023-03-29-QC-WTD-and-related-GISAID-NCBI-seqs.fasta
```

IQ-TREE maximum-likelihood tree inference with top model found using ModelFinder.

```bash
iqtree --version
```

```bash
iqtree -T 16 \
  -s 2023-03-29-QC-WTD-and-related-GISAID-NCBI-seqs.nextalign.fasta \
  -m "GTR+F+I+I+R5" \
  -o MN908947.3 \
  --prefix iqtree-2023-03-29-QC-WTD-and-related-GISAID-NCBI
```

Nextclade analysis of all sequences to get amino acid mutations in all sequences for generation of Figure 2 mutation absence/presence matrix beside phylogenetic tree.

```bash
nextclade --version
```

```bash
nextclade dataset get --name sars-cov-2 --output-dir nextclade-sars-cov-2-dataset
```

```bash
nextclade run \
  --jobs 16 \
  2023-03-29-QC-WTD-and-related-GISAID-NCBI-seqs.fasta \
  --input-dataset nextclade-sars-cov-2-dataset \
  --output-all nextclade
```
